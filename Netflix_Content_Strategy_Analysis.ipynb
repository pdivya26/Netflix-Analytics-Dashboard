Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("Netflix.csv", encoding="latin-1")
df.head()
```

Understanding Data

```python
df.shape
df.info()
```

Drop Unnecessary Columns

```python
# Drop empty unnamed columns
df = df.loc[:, ~df.columns.str.contains('Unnamed')]
df.shape
```

```python
df.head()
```

```python
df_clean = df.copy()
```

```python
print(df_clean.columns)
```

Check Missing Values

```python
df_clean.isna().sum().sort_values(ascending=False)
```

Handling Missing Values

```python
# Handle missing values
for col in ['director', 'cast', 'country', 'rating', 'duration']:
    df_clean[col] = df_clean[col].fillna('Unknown')

df_clean.isna().sum()
```

Checking Duplicate Values

```python
print("Total rows:", len(df_clean))
print("Exact Duplicate rows:", df_clean.duplicated().sum())
print("Logical Duplicate rows:", df_clean.duplicated(subset=['title'], keep=False).sum())
```

```python
df_clean[df_clean.duplicated(subset=['title'], keep=False)].sort_values('title')
```

Handling Duplicate Values

```python
df_clean = df_clean.drop_duplicates(subset=['title'], keep='first')
```

```python
print("Logical Duplicate rows:", df_clean.duplicated(subset=['title'], keep=False).sum())
```

```python
df_clean.shape
```

Checking Invalid Values

```python
for col in df_clean.columns:
    print(f"\n{col}")
    print(df_clean[col].value_counts(dropna=False).head(5))
```

```python
print(df_clean['rating'].value_counts(dropna=False).head(25))
```

```python
valid_ratings = [
    "TV-MA", "TV-14", "TV-PG", "R", "PG-13",
    "TV-Y7", "TV-Y", "PG", "TV-G", "G",
    "TV-Y7-FV", "NC-17"
]

df_clean = df_clean[df_clean['rating'].isin(valid_ratings)]
```

```python
df_clean = df_clean[~df_clean['title'].str.match(r'\d{1,2}-[A-Za-z]{3}', na=False)]
```

```python
df_clean['country'] = df_clean['country'].fillna("Unknown")
```

Feature Engineering

```python
# Parse dates
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors='coerce')

# Feature engineering
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month
df_clean['month_name'] = df_clean['date_added'].dt.month_name()
df_clean['duration_num'] = df_clean['duration'].str.extract(r'(\d+)').astype(float)
df_clean['main_genre'] = df_clean['listed_in'].str.split(',').str[0]
df_clean['main_country'] = df_clean['country'].fillna('Unknown').str.split(',').str[0]
```

```python
df_clean = df_clean[(df_clean['year_added'] < 2024) | df_clean['year_added'].isna()]
```

```python
df_clean.shape
```

```python
df_clean.columns
```

```python
print("Original:", len(df))
print("Cleaned:", len(df_clean))
print("Removed:", len(df) - len(df_clean))
```

Save Cleaned Dataset

```python
df_clean.to_csv("Netflix_Cleaned_Dataset.csv", index=False)
```
